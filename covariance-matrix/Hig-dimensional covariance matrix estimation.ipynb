# High-dimensional covariance matrix estimation

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("science")
```

```python
n = 1000
p = 500
X = np.random.randn(p, n)
```

## Sample covariance matrix

$$Q_{i,j} = \frac{1}{n-1} (X_i - \bar{X_i})  (X_j - \bar{X_j})$$

```python
Q = np.zeros((p, p))
for i in range(p):
    for j in range(p):
        xi = X[i, :]
        xj = X[j, :]
        Q[i,j] = (xi - xi.mean()) @ (xj - xj.mean())
Q = Q / (n - 1)
```

```python
np.linalg.norm(Q - np.cov(X))
```

```python
plt.matshow(Q)
plt.title("Covariance matrix Q")
plt.show()
```

$$P = \frac{1}{n-1} \sum_{i = 1}^{N} (X_i - \bar{X_i}) (X_i - \bar{X_i})^T$$

```python
P = np.zeros((p,p))
for i in range(n):
    x = X[:, i].reshape(p, 1)
    mean = X.mean(1).reshape(p, 1)
    P += (x - mean) @ (x - mean).T
P = P / (n - 1)
```

```python
np.linalg.norm(Q - np.cov(X))
```

```python
plt.matshow(P)
plt.title("Covariance matrix P")
plt.show()
```
